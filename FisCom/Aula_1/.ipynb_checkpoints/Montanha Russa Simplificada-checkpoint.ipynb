**PROBLEMA 1**

Vamos modelar os dois problemas discutidos em aula, começando pelo **problema 1**: o plano inclinado. Considere que a rampa tem $15m$ de altura e um ângulo $\beta$ de $30^{\circ}$. Vamos desconsiderar a resistência do ar nesse problema.

Veja que o código está pronto, bastando que você verifique as equações diferenciais do movimento nas direções x e y. Não se preocupe com os demais detalhes do código.

```python
get_ipython().magic('matplotlib qt5')
from math import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


#Agora que você já tem as equações diferenciais do problema 1, vamos implementá-las numa função chamada "EqDif1". 
#implemente sua função aqui: Lembre-se de que ela deve receber uma lista de soluções das equações diferenciais que você escreveu!

#parâmetros do problema
beta=radians(30) #ângulo de inclinação da trajetória
g=9.8 #aceleração da gravidade, em m/s2

def EqDif1(Solucoes, t):
    X=Solucoes[0]
    Y=Solucoes[1]
    Vx=Solucoes[2]
    Vy=Solucoes[3]    
    dXdt = Vx
    dYdt = Vy
    dVxdt = g*cos(beta)*sin(beta)
    dVydt = g*cos(beta)*cos(beta)-g
    #Vamos adicionar uma proteção aqui: 
    #quando a posição Y é negativa, zeramos as acelerações para não perdermos qual a velocidade final
    if Y < 0:
        dVxdt = 0
        dVydt = 0    
    return [dXdt, dYdt, dVxdt, dVydt]
```

```python
#Abaixo estamos criando uma lista de tempo de 0 a 5s com delta t de 1ms
from numpy import arange
#definimos o delta t: NESTE ARQUIVO, dt é menor por conta da animação
dt=0.05
#nossa lista de tempo
tempo=arange(0,5,dt)
```

```python
#Abaixo estamos descrevendo as condições iniciais do problema. Perceba que, de acordo com nossa modelagem,
# x inicial deve ser igual a zero, y inicial é a própria altura da rampa, e as velocidades Vx e Vy são nulas,
#pois nosso carrinho parte do repouso.
y_inicial=8
CD1=[0,y_inicial,0, 0]
```

```python
#Vamos agora executar o ODEINT para a obtenção da solução numérica do nosso sistema de equações diferenciais
#Depois de obter a solução, imprima a velocidade com que o carrinho chega no ponto A.

#executando o ODEINT
Solucao1=odeint(EqDif1, CD1, tempo)
#plotando a trajetória x por y
plt.plot(Solucao1[:,0], Solucao1[:,1])
plt.grid(True)
plt.show()
```

```python
#vamos imprimir a velocidade final a partir de Vx e Vy nas últimas posições da lista solução
vx_final=(Solucao1[:, 2])[-1]
vy_final=(Solucao1[:, 3])[-1]
v_final = sqrt(vx_final**2+vy_final**2)
print("A velocidade final do carrinho é: %.3f" %v_final, "m/s")
```

**PROBLEMA 2**

Vamos agora modelar o **problema 2**: o movimento circular. Estamos considerando que a velocidade inicial do carrinho é aquela obtida a partir dos resultados do problema 1. Agora que já temos as equações diferenciais do movimento circular, observe a função EqDif2. Veja  que o código também está pronto, bastando que você verifique a equação diferencial do movimento nas direções em $\theta$.

```python
#Abaixo está implementada a função "EqDif2". 

#parâmetros do modelo
r=5 #m
m=100 #kg

#equações diferenciais do movimento circular
def EqDif2(Solucoes, t):
    nao_registrou=True
    𝜃=Solucoes[0]
    𝜔=Solucoes[1]
    d𝜃dt=𝜔
    d𝜔dt=-(g*cos(𝜃))/r   
        
    return [d𝜃dt, d𝜔dt]
```

```python
#Abaixo implementamos as condições iniciais do problema. Perceba que o ângulo inicial é -120 graus!
#Observe o desenho do problema!

#condições iniciais do nosso problema
𝜃_inicial = -120
𝜔_inicial = v_final/r
CD2=[radians(𝜃_inicial),𝜔_inicial]
```

```python
#Execute o ODEINT utilizando a função EqDif2 e as condições iniciais desse problema
#A partir da solução, podemos calcular x(t), y(t) e N(t), certo? 
#N(t) seria a força normal em função do tempo.

#executando o ODEINT
Solucao2=odeint(EqDif2, CD2, tempo)
#calculando as coordenadas x(t), y(t) e N(t) a partir dos valores calculados de 𝜃(t)
x=[]
y=[]
N=[]
#neste "for", estamos calculando as coordenadas x e y em função de r e 𝜃
for i in range(0,len(Solucao2[:,0])):
    𝜃=(Solucao2[:,0])[i]
    𝜔=(Solucao2[:,1])[i]
    x.append(r*cos(𝜃))
    y.append(r*sin(𝜃))
    N.append(m*(𝜔**2)*r-m*g*sin(𝜃))

#Vamos plotar a trajetória x(t) por y(t)
plt.plot(x, y)
plt.grid(True)
plt.axis('equal')
plt.show()

#E também a força normal em função do tempo
plt.plot(tempo, N)
plt.grid(True)
plt.show()
```

```python
# Desafio para casa: A partir das duas soluções (Solucao1 e Solucao2), construa a trajetória x(t) e y(t) do movimento todo
# Como fazer para "encaixar" os pontos x(t) e y(t) do segundo problema no sistema de coordenadas do primeiro problema?
# Perceba que isso tem tudo a ver com o fato da circunferência do problema 2 tangenciar o plano inclinado no ângulo 
# inicial igual a -120 graus.

#Solução:
#vamos criar duas listas que receberão x(t) e y(t) das duas soluções
X_total=[]
Y_total=[]

X_trajetoria=[]
Y_trajetoria=[]

#vamos preencher essa lista com os valores da solucao 1
for i in range(0,len(Solucao1[:,0])):
    x=(Solucao1[:,0])[i]
    y=(Solucao1[:,1])[i]
    #atenção, só podemos incluir nessa lista os valores relativos a y>0, certo?
    if(y>0):
        X_total.append(x)
        Y_total.append(y)
        X_trajetoria.append(x)
        Y_trajetoria.append(y)
        
#precisamos agora obter a última posição da solução 1. Por que será?
X1=X_total[-1]
Y1=Y_total[-1]

caiu=False
𝜃_queda=0
𝜔_queda=0

#vamos agora implementar nosso segundo "for" para complementar a lista com a solução 2
#cuidado: x(t) e y(t) não estão no mesmo sistema de coordenadas do problema 1, pois o centro da
#circunferência está deslocado!
angulos=arange(radians(-120), radians(240), 0.01)

for 𝜃 in angulos:     
    #a partir de 𝜃 calculamos x e y
    x=r*cos(𝜃)
    y=r*sin(𝜃)
    #precisamos deslocar o centro da circunferência para a direita e para cima
    #só fazendo um desenho para entender!
    X_trajetoria.append(X1+x+r*sin(radians(-𝜃_inicial-90)))
    Y_trajetoria.append(Y1+y+r*cos(radians(-𝜃_inicial-90)))


for i in range(0,len(Solucao1[:,0])):
    𝜃=(Solucao2[:,0])[i]
    𝜔=(Solucao2[:,1])[i]
    #a partir de 𝜃 calculamos x e y da solução 2
    x=r*cos(𝜃)
    y=r*sin(𝜃)
    N=m*(𝜔**2)*r-m*g*sin(𝜃)
    if N < 0:
        caiu=True
        𝜃_queda=𝜃
        𝜔_queda=𝜔
        break
    #precisamos deslocar o centro da circunferência para a direita e para cima
    #só fazendo um desenho para entender!
    X_total.append(X1+x+r*sin(radians(-𝜃_inicial-90)))
    Y_total.append(Y1+y+r*cos(radians(-𝜃_inicial-90)))
    

#equação do lançamento oblíquo
def EqDif3(Solucoes, t):
    X=Solucoes[0]
    Y=Solucoes[1]
    Vx=Solucoes[2]
    Vy=Solucoes[3]    
    dXdt = Vx
    dYdt = Vy
    dVxdt = 0
    dVydt = -g
    #Vamos adicionar uma proteção aqui: quando a posição Y é negativa, não há mais acelerações e velocidades
    if Y < 0:
        dXdt = 0
        dYdt = 0
        dVxdt = 0
        dVydt = 0    
    return [dXdt, dYdt, dVxdt, dVydt]
    
    
    
if caiu:
    tempo3=arange(0,3, dt)
    Xi = X_total[-1]    
    Yi = Y_total[-1]
    
    Vxi =  -𝜔_queda*r*sin(𝜃_queda)
    Vyi =  𝜔_queda*r*cos(𝜃_queda)
    CI3=[Xi, Yi, Vxi, Vyi]
    Solucao3=odeint(EqDif3, CI3, tempo3)
    #vamos preencher essa lista com os valores da solucao 1
    for i in range(0,len(Solucao3[:,0])):
        x=(Solucao3[:,0])[i]
        y=(Solucao3[:,1])[i]        
        X_total.append(x)
        Y_total.append(y)   
    
    
#tudo pronto: vamos plotar as duas soluções como se fosse um único problema!
plt.plot(X_total, Y_total)
plt.grid(True)
plt.axis('equal')
plt.show()

```

```python
import matplotlib.patches as patches
from matplotlib import animation
import numpy as np

#este código faz com que o matplolib abra uma janela popup
%matplotlib qt5
  

#determina o tamanho da janela e iguala as escalas x e y
fig = plt.figure(figsize=(10,10), facecolor='gray')
plt.axis('equal')
ax = fig.add_subplot(111)

#vamos plotar a trajetória da montanha russa
ax.plot(X_trajetoria, Y_trajetoria, 'b--')
ax.grid(True)

#especificando as posições iniciais, tamanho e cor da bolinha animada
Bola = patches.Circle((0, 0), 0.3, fc='blue')

def init():
    #inclui as figuras que serão desenhadas    
    ax.add_patch(Bola)         
    return None

def animate(i):
    #animação da bolinha    
    Bola.center=(X_total[i], Y_total[i])     
    return None

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(Y_total), interval=40, blit=False)
plt.show()
```
