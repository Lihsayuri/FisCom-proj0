**PROBLEMA 1**

Vamos modelar os dois problemas discutidos em aula, come√ßando pelo **problema 1**: o plano inclinado. Considere que a rampa tem $15m$ de altura e um √¢ngulo $\beta$ de $30^{\circ}$. Vamos desconsiderar a resist√™ncia do ar nesse problema.

Veja que o c√≥digo est√° pronto, bastando que voc√™ verifique as equa√ß√µes diferenciais do movimento nas dire√ß√µes x e y. N√£o se preocupe com os demais detalhes do c√≥digo.

```python
get_ipython().magic('matplotlib qt5')
from math import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


#Agora que voc√™ j√° tem as equa√ß√µes diferenciais do problema 1, vamos implement√°-las numa fun√ß√£o chamada "EqDif1". 
#implemente sua fun√ß√£o aqui: Lembre-se de que ela deve receber uma lista de solu√ß√µes das equa√ß√µes diferenciais que voc√™ escreveu!

#par√¢metros do problema
beta=radians(30) #√¢ngulo de inclina√ß√£o da trajet√≥ria
g=9.8 #acelera√ß√£o da gravidade, em m/s2

def EqDif1(Solucoes, t):
    X=Solucoes[0]
    Y=Solucoes[1]
    Vx=Solucoes[2]
    Vy=Solucoes[3]    
    dXdt = Vx
    dYdt = Vy
    dVxdt = g*cos(beta)*sin(beta)
    dVydt = g*cos(beta)*cos(beta)-g
    #Vamos adicionar uma prote√ß√£o aqui: 
    #quando a posi√ß√£o Y √© negativa, zeramos as acelera√ß√µes para n√£o perdermos qual a velocidade final
    if Y < 0:
        dVxdt = 0
        dVydt = 0    
    return [dXdt, dYdt, dVxdt, dVydt]
```

```python
#Abaixo estamos criando uma lista de tempo de 0 a 5s com delta t de 1ms
from numpy import arange
#definimos o delta t: NESTE ARQUIVO, dt √© menor por conta da anima√ß√£o
dt=0.05
#nossa lista de tempo
tempo=arange(0,5,dt)
```

```python
#Abaixo estamos descrevendo as condi√ß√µes iniciais do problema. Perceba que, de acordo com nossa modelagem,
# x inicial deve ser igual a zero, y inicial √© a pr√≥pria altura da rampa, e as velocidades Vx e Vy s√£o nulas,
#pois nosso carrinho parte do repouso.
y_inicial=8
CD1=[0,y_inicial,0, 0]
```

```python
#Vamos agora executar o ODEINT para a obten√ß√£o da solu√ß√£o num√©rica do nosso sistema de equa√ß√µes diferenciais
#Depois de obter a solu√ß√£o, imprima a velocidade com que o carrinho chega no ponto A.

#executando o ODEINT
Solucao1=odeint(EqDif1, CD1, tempo)
#plotando a trajet√≥ria x por y
plt.plot(Solucao1[:,0], Solucao1[:,1])
plt.grid(True)
plt.show()
```

```python
#vamos imprimir a velocidade final a partir de Vx e Vy nas √∫ltimas posi√ß√µes da lista solu√ß√£o
vx_final=(Solucao1[:, 2])[-1]
vy_final=(Solucao1[:, 3])[-1]
v_final = sqrt(vx_final**2+vy_final**2)
print("A velocidade final do carrinho √©: %.3f" %v_final, "m/s")
```

**PROBLEMA 2**

Vamos agora modelar o **problema 2**: o movimento circular. Estamos considerando que a velocidade inicial do carrinho √© aquela obtida a partir dos resultados do problema 1. Agora que j√° temos as equa√ß√µes diferenciais do movimento circular, observe a fun√ß√£o EqDif2. Veja  que o c√≥digo tamb√©m est√° pronto, bastando que voc√™ verifique a equa√ß√£o diferencial do movimento nas dire√ß√µes em $\theta$.

```python
#Abaixo est√° implementada a fun√ß√£o "EqDif2". 

#par√¢metros do modelo
r=5 #m
m=100 #kg

#equa√ß√µes diferenciais do movimento circular
def EqDif2(Solucoes, t):
    nao_registrou=True
    ùúÉ=Solucoes[0]
    ùúî=Solucoes[1]
    dùúÉdt=ùúî
    dùúîdt=-(g*cos(ùúÉ))/r   
        
    return [dùúÉdt, dùúîdt]
```

```python
#Abaixo implementamos as condi√ß√µes iniciais do problema. Perceba que o √¢ngulo inicial √© -120 graus!
#Observe o desenho do problema!

#condi√ß√µes iniciais do nosso problema
ùúÉ_inicial = -120
ùúî_inicial = v_final/r
CD2=[radians(ùúÉ_inicial),ùúî_inicial]
```

```python
#Execute o ODEINT utilizando a fun√ß√£o EqDif2 e as condi√ß√µes iniciais desse problema
#A partir da solu√ß√£o, podemos calcular x(t), y(t) e N(t), certo? 
#N(t) seria a for√ßa normal em fun√ß√£o do tempo.

#executando o ODEINT
Solucao2=odeint(EqDif2, CD2, tempo)
#calculando as coordenadas x(t), y(t) e N(t) a partir dos valores calculados de ùúÉ(t)
x=[]
y=[]
N=[]
#neste "for", estamos calculando as coordenadas x e y em fun√ß√£o de r e ùúÉ
for i in range(0,len(Solucao2[:,0])):
    ùúÉ=(Solucao2[:,0])[i]
    ùúî=(Solucao2[:,1])[i]
    x.append(r*cos(ùúÉ))
    y.append(r*sin(ùúÉ))
    N.append(m*(ùúî**2)*r-m*g*sin(ùúÉ))

#Vamos plotar a trajet√≥ria x(t) por y(t)
plt.plot(x, y)
plt.grid(True)
plt.axis('equal')
plt.show()

#E tamb√©m a for√ßa normal em fun√ß√£o do tempo
plt.plot(tempo, N)
plt.grid(True)
plt.show()
```

```python
# Desafio para casa: A partir das duas solu√ß√µes (Solucao1 e Solucao2), construa a trajet√≥ria x(t) e y(t) do movimento todo
# Como fazer para "encaixar" os pontos x(t) e y(t) do segundo problema no sistema de coordenadas do primeiro problema?
# Perceba que isso tem tudo a ver com o fato da circunfer√™ncia do problema 2 tangenciar o plano inclinado no √¢ngulo 
# inicial igual a -120 graus.

#Solu√ß√£o:
#vamos criar duas listas que receber√£o x(t) e y(t) das duas solu√ß√µes
X_total=[]
Y_total=[]

X_trajetoria=[]
Y_trajetoria=[]

#vamos preencher essa lista com os valores da solucao 1
for i in range(0,len(Solucao1[:,0])):
    x=(Solucao1[:,0])[i]
    y=(Solucao1[:,1])[i]
    #aten√ß√£o, s√≥ podemos incluir nessa lista os valores relativos a y>0, certo?
    if(y>0):
        X_total.append(x)
        Y_total.append(y)
        X_trajetoria.append(x)
        Y_trajetoria.append(y)
        
#precisamos agora obter a √∫ltima posi√ß√£o da solu√ß√£o 1. Por que ser√°?
X1=X_total[-1]
Y1=Y_total[-1]

caiu=False
ùúÉ_queda=0
ùúî_queda=0

#vamos agora implementar nosso segundo "for" para complementar a lista com a solu√ß√£o 2
#cuidado: x(t) e y(t) n√£o est√£o no mesmo sistema de coordenadas do problema 1, pois o centro da
#circunfer√™ncia est√° deslocado!
angulos=arange(radians(-120), radians(240), 0.01)

for ùúÉ in angulos:     
    #a partir de ùúÉ calculamos x e y
    x=r*cos(ùúÉ)
    y=r*sin(ùúÉ)
    #precisamos deslocar o centro da circunfer√™ncia para a direita e para cima
    #s√≥ fazendo um desenho para entender!
    X_trajetoria.append(X1+x+r*sin(radians(-ùúÉ_inicial-90)))
    Y_trajetoria.append(Y1+y+r*cos(radians(-ùúÉ_inicial-90)))


for i in range(0,len(Solucao1[:,0])):
    ùúÉ=(Solucao2[:,0])[i]
    ùúî=(Solucao2[:,1])[i]
    #a partir de ùúÉ calculamos x e y da solu√ß√£o 2
    x=r*cos(ùúÉ)
    y=r*sin(ùúÉ)
    N=m*(ùúî**2)*r-m*g*sin(ùúÉ)
    if N < 0:
        caiu=True
        ùúÉ_queda=ùúÉ
        ùúî_queda=ùúî
        break
    #precisamos deslocar o centro da circunfer√™ncia para a direita e para cima
    #s√≥ fazendo um desenho para entender!
    X_total.append(X1+x+r*sin(radians(-ùúÉ_inicial-90)))
    Y_total.append(Y1+y+r*cos(radians(-ùúÉ_inicial-90)))
    

#equa√ß√£o do lan√ßamento obl√≠quo
def EqDif3(Solucoes, t):
    X=Solucoes[0]
    Y=Solucoes[1]
    Vx=Solucoes[2]
    Vy=Solucoes[3]    
    dXdt = Vx
    dYdt = Vy
    dVxdt = 0
    dVydt = -g
    #Vamos adicionar uma prote√ß√£o aqui: quando a posi√ß√£o Y √© negativa, n√£o h√° mais acelera√ß√µes e velocidades
    if Y < 0:
        dXdt = 0
        dYdt = 0
        dVxdt = 0
        dVydt = 0    
    return [dXdt, dYdt, dVxdt, dVydt]
    
    
    
if caiu:
    tempo3=arange(0,3, dt)
    Xi = X_total[-1]    
    Yi = Y_total[-1]
    
    Vxi =  -ùúî_queda*r*sin(ùúÉ_queda)
    Vyi =  ùúî_queda*r*cos(ùúÉ_queda)
    CI3=[Xi, Yi, Vxi, Vyi]
    Solucao3=odeint(EqDif3, CI3, tempo3)
    #vamos preencher essa lista com os valores da solucao 1
    for i in range(0,len(Solucao3[:,0])):
        x=(Solucao3[:,0])[i]
        y=(Solucao3[:,1])[i]        
        X_total.append(x)
        Y_total.append(y)   
    
    
#tudo pronto: vamos plotar as duas solu√ß√µes como se fosse um √∫nico problema!
plt.plot(X_total, Y_total)
plt.grid(True)
plt.axis('equal')
plt.show()

```

```python
import matplotlib.patches as patches
from matplotlib import animation
import numpy as np

#este c√≥digo faz com que o matplolib abra uma janela popup
%matplotlib qt5
  

#determina o tamanho da janela e iguala as escalas x e y
fig = plt.figure(figsize=(10,10), facecolor='gray')
plt.axis('equal')
ax = fig.add_subplot(111)

#vamos plotar a trajet√≥ria da montanha russa
ax.plot(X_trajetoria, Y_trajetoria, 'b--')
ax.grid(True)

#especificando as posi√ß√µes iniciais, tamanho e cor da bolinha animada
Bola = patches.Circle((0, 0), 0.3, fc='blue')

def init():
    #inclui as figuras que ser√£o desenhadas    
    ax.add_patch(Bola)         
    return None

def animate(i):
    #anima√ß√£o da bolinha    
    Bola.center=(X_total[i], Y_total[i])     
    return None

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(Y_total), interval=40, blit=False)
plt.show()
```
