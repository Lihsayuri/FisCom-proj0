**1) Implementação**

A partir do sistema do saltador definido em aula, plote em um mesmo gráfico a altura do anel ($y_a$) e da barra ($y_b$) em função do tempo com as seguintes condições iniciais:
- $y_a(0) = 0$
- $y_b(0) = h$
- $v_a(0) = 0$
- $v_b(0) = 0$

Considere os seguintes valores de aceleração da gravidade ($g$), massa do anel ($m_b$), massa da barra ($m_b$), altura da barra ($h$), raio do anel ($r$), comprimento natural (quando não esticado) do elástico ($l_0$) e constante elástica do elástico ($k$):
- $g = 9,81$ m/s$^2$
- $m_a = 0,01$ kg
- $m_b = 0,008$ kg
- $h = 0,1$ m
- $r_a = 0,05$ m
- $r_b = 0,005$ m
- $l_0 = 0,04$ m
- $k = 3$ N/m

**Observação:** execute a simulação por 3 segundos.

```python
# Importa bibliotecas
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# SEU CÓDIGO

#ya[0]=0 
#yb[0]=h
#va[0]=0
#vb[0]=0

g=9.81  #m/s2
ma=0.01 #kg
mb=0.008 #kg
h=0.1 #m
ra=0.05 #m
rb=0.005 #m
l0=0.04 #m
k=3 #N/m

def modelo (x,t):
    ya=x[0]
    yb=x[1]
    va=x[2]
    vb=x[3]
    RA= (ra-rb)**2
    YB= (yb-ya)**2
    dyadt= va
    dybdt= vb
    dvadt=(3*k*((RA+YB)**0.5-l0)*((yb-ya)/(RA+YB)**0.5)/ma)-g
    dvbdt=(-3*k*((RA+YB)**0.5-l0)*((yb-ya)/(RA+YB)**0.5)/mb)-g
    if (yb<=h and dvbdt<0):
        dybdt=0
        dvbdt=0
    if (yb<h):
        dybdt=0
        dvbdt=0
    dxdt=[dyadt,dybdt,dvadt,dvbdt]
    return dxdt

delta_t= 1e-2
lista_t=np.arange(0,3,delta_t)
ya_0=0
yb_0=h
va_0=0
vb_0=0
x_0=[ya_0,yb_0,va_0,vb_0]

x=odeint(modelo,x_0,lista_t)
lista_ya=x[:,0]
lista_yb=x[:,1]

plt.plot(lista_t,lista_ya,label='valores da altura do anel')
plt.plot(lista_t,lista_yb, label= 'valores da altura da barra')
plt.xlabel('valores de tempo em segundos')
plt.ylabel('valores das alturas em metros')
plt.legend()
plt.grid(True)
plt.show()


```

**2) Uma forma de validação**

Uma forma interessante de validar a sua implementação é verificar a variação da energia do sistema. A rigor, não estaremos validando o modelo em comparação com um caso real, mas verificaremos se ele é fisicamente coerente.

Como não foi considerado nenhum tipo de dissipação de energia, a energia mecânica ($E_m$) deve permanecer constante. Sabendo que a mesma é a somatória das energias cinética ($E_c$), potencial gravitacional ($E_{pg}$) e potencial elática ($E_{pe}$):
- $E_m = E_c+E_{pg}+E_{pe}$

E que cada uma delas é dada por:
- $E_c = \frac{1}{2} m_a \cdot v_a^2 + \frac{1}{2} m_b \cdot v_b^2 $
- $E_{pg} = m_a \cdot g \cdot y_a + m_b \cdot g \cdot y_b$
- $E_{pe} = 3 \left[ \frac{1}{2} k {\left( \sqrt{(r_b-r_a)^2+(y_b-y_a)^2}-l_0 \right)}^2 \right] $

Plote em um mesmo gráfico todas as energias em função do tempo e verifique se a energia mecânica de fato permanece constante.

```python
# SEU CÓDIGO

Ec=[]
Epg=[]
Epe=[]
Em=[]

for i in range(len(lista_t)):
    lista_ya=x[:,0]
    lista_yb=x[:,1]
    lista_va=x[:,2]
    lista_vb=x[:,3]
    RB= ((rb-ra)**2+(lista_yb[i]-lista_ya[i])**2)**0.5
    Ec.append(0.5*ma*(lista_va[i])**2+0.5*mb*(lista_vb[i])**2)
    Epg.append(ma*g*lista_ya[i]+mb*g*lista_yb[i])
    Epe.append(3*(0.5*k*(RB-l0)**2))
    Em.append(Ec[i]+Epg[i]+Epe[i])

plt.plot(lista_t, Ec, label="Energia cinética")
plt.plot(lista_t, Epg, label="Energia potencial gravitacional")
plt.plot(lista_t, Epe, label="Energia potencial elástica")
plt.plot(lista_t, Em, label="Energia Mecânica")
plt.xlabel('Tempo (s)')
plt.ylabel('Energia (J)')
plt.legend()
plt.grid(True)
plt.show()







```
